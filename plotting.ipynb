```python
import matplotlib.pyplot as plt
import pandas as pd
#Sample data
data = {'subjects':['CNS','AI','OOAD','ST','HCI','CDC'],'NO OF CLASSES':[10,15,9,5,4,1]}
df = pd.DataFrame(data)
#create a barchart using matplotlib
plt.bar(df['subjects'],df['NO OF CLASSES'],width=0.5,color='magenta')
#title
plt.title('Subjects and class details')
#labels
plt.xlabel('subjects')
plt.ylabel('class details')
#legend
plt.legend('s')
#Display the chart
plt.show()
```
